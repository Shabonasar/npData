## прохождение пробки через компенсатор

Рассмотрим прохождение пробки через компенсатор. Для этого выберем момент времени когда через него проходит пробка и посмотрим на него в нескольких масштабах - исходные данные и данные осредненные через 100 мс

готовим данные и строим обзорные графики по полученным замерам

```python
t4s=t4.df.resample('100ms').mean()
t5s=t5.df.resample('100ms').mean()
t6s=t6.df.resample('100ms').mean()
st4 =pd.Timestamp('2017-09-16 08:22:00') 
end4=pd.Timestamp('2017-09-16 16:17:00')
st5 =pd.Timestamp('2017-09-16 08:20:00 ')
end5=pd.Timestamp('2017-09-16 16:17:00')
st6 =pd.Timestamp('2017-09-16 08:16:00')
end6=pd.Timestamp('2017-09-16 16:17:00')
t4s[st4:end4].plot()
t5s[st5:end5].plot()
t6s[st6:end6].plot()
```

Смотрим на данные за короткий промежуток времени за который через компенсатор прошла пробка

```python
st = pd.Timestamp('2017-09-16 06:08:12')
en = pd.Timestamp('2017-09-16 06:08:30')
t4.df[st:en].aX.plot()
t5.df[st:en].aX.plot()
t6.df[st:en].aX.plot()
```
